```python
import numpy as np
from numpy.linalg import norm
class Kmeans:
    def __init__(self, n_clusters, max_iter=100, random_state=101):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.random_state = random_state
```

```python
    def initializ_centroids(self, data):
        np.random.RandomState(self.random_state)
        random_idx = np.random.permutation(data.shape[0])
        centroids = data[random_idx[:self.n_clusters]]
        return centroids
```

```python
    def compute_centroids(self, data, labels):
        centroids = np.zeros((self.n_clusters, data.shape[1]))
        for k in range(self.n_clusters):
            centroids[k, :] = np.mean(data[labels == k, :], axis=0)
        return centroids
```

```python
def compute_distance(self, data, centroids):
        distance = np.zeros((data.shape[0], self.n_clusters))
        for k in range(self.n_clusters):
            row_norm = norm(data - centroids[k, :], axis=1)
            distance[:, k] = np.square(row_norm)
        return distance
```

```python
    def find_closest_cluster(self, distance):
        return np.argmin(distance, axis=1)
```

```python
def compute_sse(self, data, labels, centroids):
        distance = np.zeros(data.shape[0])
        for k in range(self.n_clusters):
            distance[labels == k] = norm(data[labels == k] - centroids[k], axis=1)
        return np.sum(np.square(distance))
```

```python
    def fit(self, data):
        self.centroids = self.initializ_centroids(data)
        for i in range(self.max_iter):
            old_centroids = self.centroids
            distance = self.compute_distance(data, old_centroids)
            self.labels = self.find_closest_cluster(distance)
            self.centroids = self.compute_centroids(data, self.labels)
            if np.all(old_centroids == self.centroids):
                break
        self.error = self.compute_sse(data, self.labels, self.centroids)
```

```python
    def predict(self, data):
        distance = self.compute_distance(data, old_centroids)
        return self.find_closest_cluster(distance)
```

```python
import pandas as pd
import random
import re
%matplotlib inline
import numpy as np
from scipy import stats, integrate
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import random
from numpy.random import randn

mean1=[0,0]
cov=[[1,0],[0,1]]
mean2=[3,0]
mean3=[0,3]
wt=[0.2, 0.5, 0.3]

x1 = np.random.multivariate_normal(mean1, cov, (100))
x2 = np.random.multivariate_normal(mean2, cov, (250))
x3 = np.random.multivariate_normal(mean3, cov, (150))
x=np.vstack((x1,x2,x3))

x.shape

 
def pick_mu(k):
    copy_1=x
    np.random.shuffle(copy_1)
    mu=copy_1[:k]
    return mu

def kmeans(k):
    c=np.zeros(500).astype(int)
    c=c.astype(int)
    L=[]
    mu=pick_mu(k)
    for t in range(20):
        

        for i in range(x.shape[0]):
            dist=[]
            for k1 in range(k):
                dist.append(np.sqrt(np.sum((x[i]-mu[k1])**2)))
            c[i]=np.argmin(dist)+1

        for k1 in range(k):
            nk=0
            for i in range(x.shape[0]):
                if c[i]==k1+1:
                    nk=nk+1
            val=np.zeros((1,2))
            for i in range(x.shape[0]):
                if c[i]==k1+1:
                    val=np.add(val,x[i])
        
            mu[k1]=val/nk
            
        sum1=0
        for i in range(x.shape[0]):
            for k1 in range(k):
                if c[i]==k1+1:
                    sum1=sum1+(np.sum((x[i]-mu[k1])**2))
        L.append(sum1)
    return L,mu,c

```

```python
x_axis2 = np.arange(1, 21)
import matplotlib.pyplot as plt
plt.figure(figsize=(9,6))
l2,mu2,c2=kmeans(2)
plt.plot(x_axis2,l2,label="k=2")
l3,mu3,c3=kmeans(3)
plt.plot(x_axis2,l3,label="k=3")
l4,mu4,c4=kmeans(4)
plt.plot(x_axis2,l4,label="k=4")
l5,mu5,c5=kmeans(5)
plt.plot(x_axis2,l5,label="k=5")
plt.xticks(np.arange(min(x_axis2), max(x_axis2)+1, 1.0))
plt.legend(loc='upper right')
plt.title("Objective Training Function L per iteration ")
plt.xlabel("Iterations ")
plt.ylabel("Objective Training Function")
plt.show()
```
